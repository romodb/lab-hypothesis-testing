```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
pokemon = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
pokemon
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# Filtramos:
t_dragon = pokemon[pokemon['Type 1'] == 'Dragon']['HP']
t_cesped = pokemon[pokemon['Type 1'] == 'Grass']['HP']

# Hacemos T Student:
t_stat, p_value = st.ttest_ind(t_dragon, t_cesped, alternative='greater')

# Obtenemos los resultados: 
print(f"El valor del test es {t_stat} y el valor de p es  {p_value}")

# Evaluamos el valor p: 
if p_value < 0.5:
    print("Se rechaza la hipotesis al tener el tipo dragon un Hp mayor al tipo planta.")
else:
    print("No se rechaza la hipotesis, pero por falta de datos.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
pokemon.info()
```

```python
# Filtrado: 
t_legendario = pokemon[pokemon['Legendary'] == True]
n_t_lengendario = pokemon[pokemon['Legendary'] == False]

# Atributos a comparar: 
atributos = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Haceos t_student para cada atributo:
ajustes = 0.05 / len(atributos)
for atributo in atributos: 
    stadist_legendario = t_legendario[atributo]
    stadist_no_legendario = n_t_lengendario[atributo]

# Hacemos t-student:
    t_stat, p_value = st.ttest_ind(stadist_legendario, stadist_no_legendario, equal_var=False)  

# Revisamos valores:
print(f"Comparacion para {atributos}, el valor de t_student es {t_stat}, para p es {p_value}")

# Veamos si aceptamos o rechazmos la hipotesis: 
if p_value < ajustes:
    print(f"Se rechaza la hipotesis debido que hay una diferencia significativa entre {atributos} de los legendarios y los no legendarios.")
else: 
    print("No se rechaza la hipotesis al no haber una diferencia signficativa entre categorías.")
```

In this challenge, we will be working with california-housing data. The data can be found here:

```python
distancias = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
distancias.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Se declaran las distancias:
coor_escuela = (-118, 34)
coor_hosp = (-122, 37)

# Se crea la función para calcular la distancia euclidea:
def dist_eu (x1, y1, x2, y2):
    return np.sqrt((x2 -x1)**2 + (y2 - y1)**2)

# Calculamos las distancias de cada casa con el hospital y el colegio: 
distancias['d_hospital'] = distancias.apply(lambda row: dist_eu(row['X'], row['Y'], coor_hosp[0], coor_hosp[1]), axis=1)
distancias['d_hospital'] = distancias.apply(lambda row: dist_eu(row['X'], row['Y'], coor_escuela[0], coor_escuela[1]), axis=1)

# Clasificamos las casas como "lejos" o "cerca":
distancias['cerca'] = ((distancias['d_hospital'] < 0.5) | (distancias['d_escuela'] < 0.5))

# Dividimos el dataset en dos: 
cerca = distancias[distancias['Cerca'] == True]['Price']
lejos = distancias[distancias['Lejos'] == True]['Price']

# ##
t_stat, p_value = st.ttest_ind(cerca, lejos, equal_var=False)
print(f"Valor de t {t_stat}, valor de p {p_value}")

if p_value < 0.5:
    print("Se rechaza la hipotesis nula.")
else: 
    print("No se rechaza la hipotesis.")

```
